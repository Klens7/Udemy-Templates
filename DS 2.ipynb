##### Choropleth Maps

```python
# imports and everything to be working offline

import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
# set up everything so that the figures show up in the notebook

init_notebook_mode(connected=True)
```

```python
import pandas as pd
```

Now we need to begin to build our data dictionary. Easiest way to do this is use the dict() function 

type ='choropleth
locations=list of states
locationmode='USA-states'
colorscale=

Either a predefined string:

'pairs'  'Greys' 'Greens'  'Blurred'    'Hot'  'Picnic'  

```python
# dict to cast a list to a dictionary

data = dict(type = 'choropleth', 
            # state abbrevation codes
            locations = ['AZ', 'CA', 'NY'],
            
            # to let plotly know its US, can go down to county level
            locationmode = 'USA-states',
            colorscale= 'Jet',
            
            # text is what hovers over the locations, index sequence as above states
            text=['text1', 'text2', 'text3'],
            
            # z is values in the color scale
            z = [1.0,2.0,3.0],
            
            # dictionary of key title and whatever color bar title is
            colorbar= {'title':'Colorbar Title'})
```

```python
data
```

```python
# then create the layout we need
# dict to cast a list to a dictionary
# nested dictionary

layout = dict(geo = {'scope':'usa'})
```

```python
choromap = go.Figure(data = [data],layout=layout)
```

```python
iplot(choromap)
```

### Arriculture Data Graph

```python
df = pd.read_csv('2011_US_AGRI_Exports')
```

```python
df.head()
```

```python
## data variable and layout variable needed, like above

data = dict(type='choropleth',
           colorscale='ylorbr',
            locations = df['code'],
            locationmode = 'USA-states',
            z = df['total exports'],
            text = df['text'],
            marker = dict(line = dict(color = 'rgb(255,255,255)',width=2)), # can do 12,12,12 and width = 4 for dark borders 
            colorbar = {'title': 'Millions USD'}
            
           )
```

```python
layout = dict(title = '2011 US Argriculture Exports by State',
              geo = dict(scope='usa',showlakes = True, lakecolor='rgb(85,173,240)'))
```

```python
layout
```

```python
choromap2 = go.Figure(data = [data],layout=layout)
```

```python
iplot(choromap2)
```

```python
import numpy as np

x = np.arange(0,100)

y = x * 2

z = x**3
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y, color='red')
ax2.plot(y,x)
```

```python
fig,axes = plt.subplots(1,2, figsize=(12,2))

axes[0].plot(x,y, color='blue', lw=3,ls='--')
axes[1].plot(x,z,color='red',lw=3)
```

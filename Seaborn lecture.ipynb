## Distribution plots

distribution plots allows us to view the distribution of a dataset

```python
import seaborn as sns
```

```python
%matplotlib inline

# so we can see it in jupyter notebook
```

```python
# Seaborn has built in data set we can load

# lets use the data set TIPS 
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# distribtuion of uni-variate  (one variable)

sns.distplot(tips['total_bill'], bins=30)

# kde kernal density estimation - the line below
# can remove by arguement kde=False
# can add bins=30
```

```python
# joint plot to match up two dist plots
# pass in x variable (string on column name), y variable (string of column name for comparison), and dataset
# comparing total bill as x and tips as y

# gives you a scatter plot and 2 histograms

sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
# 'kind = scatter' is default
# can add arguement kind='hex' or 'reg' or 'kde' for hexagons colored, for regression line, for dimensional kde 
```

```python
# pair plot
# joint plot for all possible combinations in the dataframe
# pass in dataframe
# for all numerical column values

sns.pairplot(tips, hue='sex', palette='coolwarm')
# can add hue='sex' (the column we want)  arguement for catagorical columns (guy / girl)
# palette='coolwarm' for color arguement
```

```python
# rug plots
# draws dash for all point on this variable

sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

## Categorical Plots

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')

tips.head()
```

```python
# barplot
# general plot to aggregate categorical data

# visialization of group by action
# x = 'catagorical column' like male/female 
# y = 'numeric column'

sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)

# estimator arguement (who has more variance) (stand deviation) per category
##   import numpy as np

# total bill for males slightly higher than females
```

```python
import numpy as np
```

```python
# count plot
# only set x column
# like pandas.count
# how many males and females

sns.countplot(x='sex', data=tips)
```

```python
# box plots 
# somestimes called box and wisker plot

sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')

# can split more with arguement hue='smoker' category
```

```python
# violin plot

sns.violinplot(x='day',y='total_bill', data=tips, hue='sex', split=True)
```

```python
# strip plot
# scatter plot based off category

sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
# swarm plot
# a strip plot with points that don't overlap
# shows all points

sns.swarmplot(x='day', y='total_bill', data=tips)
```

## Matrix Plots


plots (like heat maps)


```python
import seaborn as sns
%matplotlib inline


# 2 built in datasets
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

tips.head()
```

```python
flights.head()
```

```python
# matrix form we need data in

tc = tips.corr()
```

```python
sns.heatmap(tc, annot = True)

# added annotate arguement 
# can add cmap='coolwarm' for colors
```

```python
flights

# need to convert to matrix form
```

```python
fp = flights.pivot_table(index='month',columns='year', values='passengers')

# this is for matrix form
# set it as a variable

fp
```

```python
# heat map to visual

sns.heatmap(fp, cmap='magma',linecolor='white',linewidth=2)

# added arguements cmap,linecolor,linewidth
```

```python
# clustered version of heatmap

sns.clustermap(fp)
```

## Regression Plots

seaborn has many built in capabilites for regression plots

great for machine learning

(LM plot function) Linear Model plot function - displays linear models with seaborn

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y= 'tip', data=tips)

# can add hue='sex' to break them out
# can pass in matplot lib style parameters

# scatter plot with a linear fit on top of it 
```

```python
sns.lmplot(x='total_bill', y= 'tip', data=tips, hue='sex', markers=['o','v'],
          scatter_kws ={'s':100})

# a call to matplotlib
# matplotlib is under the hood 
```

```python
# grid by column and row
sns.lmplot(x='total_bill', y= 'tip', data=tips, col='day',row='time',aspect=0.6)
```
